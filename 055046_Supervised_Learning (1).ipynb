#### **5.5. Machine Learning Models**

- **Logistic Regression (LR)**:
  Logistic regression is a classification algorithm that models the probability of a binary outcome based on one or more predictor variables using a logistic function.
  
- **Support Vector Machines (SVM)**:
  SVM finds a hyperplane that best separates the data into classes by maximizing the margin between the closest points of the classes (support vectors).

- **Stochastic Gradient Descent (SGD)**:
  An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.

- **Decision Trees**:
  A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.

- **K-Nearest Neighbors (KNN)**:
  A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
  
- **Naive Bayes (NB)**:
  A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.

- **Bagging (Bootstrap Aggregating)**:
  Combines predictions from multiple models (e.g., Random Forest) to reduce variance and improve accuracy.

- **Boosting**:
  Combines weak learners sequentially to form a strong learner by minimizing errors iteratively (e.g., Extreme Gradient Boosting - XGBoost).

#### **5.6. Model Performance Metrics**

- **Confusion Matrix**:
  - Sensitivity (Recall): True Positive Rate
  - Specificity: True Negative Rate
  - Accuracy: Overall correctness
  - Precision: Positive Predictive Value
  - F1-Score: Harmonic mean of Precision and Recall
  - AUC: Area Under the ROC Curve

- **K-Fold Cross-Validation**: Splits data into k subsets; trains on k-1 and tests on the remaining one iteratively to ensure robust performance evaluation.

- **Model Run Statistics**:
  - Time Taken

### **8. Managerial Insights and Recommendations **

#### **8.1 Preprocessing of Data**

##### **Recommended Treatment of Missing Data**
- **Insight**: The dataset contains no missing values, ensuring a more streamlined analysis process. This indicates that data collection and input mechanisms are robust and there is no immediate concern regarding missing data.
- **Recommendation**: Although no missing values were found, it's recommended to implement data imputation strategies like mean, median, or mode imputation for future datasets in case of missing values. Additionally, developing a real-time monitoring system to detect and handle missing values automatically would ensure consistent data integrity.

##### **Recommended Numerical Encoding of Non-Numeric Categorical Data**
- **Insight**: The dataset utilized **Ordinal Encoding** for non-numeric categorical data. This encoding works well when the categorical variables have an inherent order (e.g., "Payment_Terms").
- **Recommendation**: For **nominal** variables (such as "Country", "Product", "Category"), **One-Hot Encoding** should be applied. This prevents assumptions of ordinality (e.g., assuming one country is 'higher' than another) and helps the models better interpret independent categories without creating spurious relationships. For ordinal data like "Payment_Terms", ordinal encoding remains suitable.  
  - Implement checks to ensure that encoding techniques align with the data type to avoid biased model assumptions.

##### **Recommended Treatment of Non-Categorical Numeric Data having Non-Uniform Scales & Outliers**
- **Insight**: Non-categorical numerical data such as **Quantity**, **Value**, and **Weight** were preprocessed using **Min-Max Scaling**, which normalizes the data within a specific range (0-1). While this method is useful for many models, it could be impacted by outliers, as Min-Max scaling is sensitive to extreme values.
- **Recommendation**: Consider alternatives such as **Robust Scaling** or **Log Transformation** to deal with highly skewed distributions or extreme outliers. **Robust Scaling** is better for datasets with outliers because it uses the median and interquartile range to scale the data, minimizing the effect of outliers. If outliers represent valid information, they should not be removed but managed accordingly.
  - Additionally, evaluate **Z-score Normalization** for standardized scaling (mean = 0, standard deviation = 1) for algorithms like SVM that are sensitive to feature scaling.

##### **Recommended Split [70:30 | 75:25 | 80:20] of Sample Data into Training & Testing Sets**
- **Insight**: The dataset was split into training and testing sets. A well-established practice is to allocate around 70% for training and 30% for testing, ensuring sufficient data to train the model while retaining a fair test set for validation.
- **Recommendation**: Given the nature of the dataset, an **80:20 split** can be considered for **more balanced training** data if the dataset is large. For imbalanced classes (e.g., a disproportionate number of imports vs exports), **Stratified Sampling** should be applied to ensure that the class distributions in both training and testing sets are representative.
  - Consider experimenting with **cross-validation** (e.g., 5-fold or 10-fold) to assess model stability and robustness by splitting the data into multiple subsets.

##### **Insight on its Uniqueness, Usability, and Suitability Based on the Model Characteristics**

1. **Logistic Regression**:
   - **Uniqueness**: Fast and interpretable, especially for binary classification tasks.
   - **Usability**: Suitable for real-time applications due to its speed and simplicity.
   - **Suitability**: Best suited for datasets where there is a **linear relationship** between the features and the outcome. Less suitable for non-linear patterns without feature transformations.

2. **K-Nearest Neighbors (KNN)**:
   - **Uniqueness**: Simple to implement and intuitively understandable. Works well with non-linear decision boundaries.
   - **Usability**: Best for classification tasks where the class decision depends on local data relationships. However, becomes computationally expensive as the dataset grows.
   - **Suitability**: Best for small-to-medium datasets. Works efficiently when data points are **spatially clustered**.

3. **Decision Tree**:
   - **Uniqueness**: Provides clear decision rules, easily interpretable, and works well with mixed data types (numeric and categorical).
   - **Usability**: Works well for data that requires interpretability and clear decision logic.
   - **Suitability**: Prone to overfitting, especially with many categorical features. Requires tuning for optimal performance (e.g., through pruning, max depth, etc.).

4. **Random Forest**:
   - **Uniqueness**: Uses multiple decision trees to reduce overfitting. Provides a more robust model than a single decision tree.
   - **Usability**: Very powerful but at the cost of higher computational complexity.
   - **Suitability**: Best suited for high-dimensional datasets and when accuracy is prioritized over interpretability.

5. **Support Vector Machine (SVM)**:
   - **Uniqueness**: Excellent for datasets with complex decision boundaries.
   - **Usability**: Suitable for binary classification tasks with high-dimensional data. Can struggle with large datasets and is computationally expensive.
   - **Suitability**: Best used in situations where there is a clear margin of separation between classes. Scaling of features is critical for SVMs to perform well.

```python
def summarize_numerical(my_sample, variables):
    summary = {}

    for col in variables:
        if col in ks46_sample[ks46_non_cat_vars]:  # Check if the column exists in the DataFrame
            summary[col] = {
                'Mode': my_sample[col].mode()[0],
                'Range': my_sample[col].max() - my_sample[col].min(),
                'Skewness': my_sample[col].skew() ,
                'Kurtosis': my_sample[col].kurtosis()
            }
    return summary
```

```python
# Coefficient of Variation
def coefficient_of_variation(data):
    mean = data.mean()
    std = data.std()
    cv = (std / mean) * 100
    return cv

cv_results = {}

for col in ks46_non_cat_vars:
    cv_results[col] = coefficient_of_variation(ks46_sample[col])

# Display the results
print("Coefficient of Variation (%):")
for col, cv in cv_results.items():
    print(f"{col}: {cv:.2f}%")
```

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in ks46_non_cat_vars:
    ci_results[col] = confidence_interval(ks46_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

ks46_contingency_table = pd.crosstab(ks46_sample['Import_Export'], ks46_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(ks46_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
ks46_alpha = 0.05  # Significance level

if p < ks46_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    ks46_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    ks46_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    ks46_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    ks46_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    ks46_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return ks46_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in ks46_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(ks46_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(ks46_non_cat_vars) > 1:
        for i in range(len(ks46_non_cat_vars)):
            for j in range(i+1, len(ks46_non_cat_vars)):
                col1, col2 = ks46_non_cat_vars[i], ks46_non_cat_vars[j]
                t_test_result = t_test_correlation(ks46_sample[col1], ks46_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(ks46_model, ks46_X, ks46_y): # Function parameters renamed
    tracemalloc.start()
    ks46_model.fit(ks46_X, ks46_y)
    ks46_current, ks46_peak = tracemalloc.get_traced_memory() # Variables renamed
    tracemalloc.stop()
    return ks46_peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
ks46_log_reg_model = LogisticRegression(random_state=55046) # Variable renamed
ks46_log_reg_memory = get_model_memory_usage(ks46_log_reg_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"Logistic Regression memory usage: {ks46_log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
ks46_svm_model = SVC(random_state=55046) # Variable renamed
ks46_svm_memory = get_model_memory_usage(ks46_svm_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"SVM memory usage: {ks46_svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
ks46_sgd_model = SGDClassifier(random_state=55046) # Variable renamed
ks46_sgd_memory = get_model_memory_usage(ks46_sgd_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"SGD memory usage: {ks46_sgd_memory:.2f} MB")

# Decision Tree
ks46_dt_model = DecisionTreeClassifier(random_state=55046) # Variable renamed
ks46_dt_memory = get_model_memory_usage(ks46_dt_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"Decision Tree memory usage: {ks46_dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
ks46_knn_model = KNeighborsClassifier(n_neighbors=5) # Variable renamed
ks46_knn_memory = get_model_memory_usage(ks46_knn_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"KNN memory usage: {ks46_knn_memory:.2f} MB")

# Naive Bayes
ks46_nb_model = GaussianNB() # Variable renamed
ks46_nb_memory = get_model_memory_usage(ks46_nb_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"Naive Bayes memory usage: {ks46_nb_memory:.2f} MB")

# Random Forest
ks46_rf_model = RandomForestClassifier(random_state=55046) # Variable renamed
ks46_rf_memory = get_model_memory_usage(ks46_rf_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"Random Forest memory usage: {ks46_rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
ks46_xgb_model = XGBClassifier(random_state=55046) # Variable renamed
ks46_xgb_memory = get_model_memory_usage(ks46_xgb_model, ks46_X_train, ks46_y_train) # Variables renamed
print(f"XGBoost memory usage: {ks46_xgb_memory:.2f} MB")
```
